### Data Cleaning

```python
# Check missing Values 
df.isnull().sum()
```

```python
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)
```

```python
# filling missing values in numeric columns with median
df['Attendance (%)'] = df['Attendance (%)'].fillna(df['Attendance (%)'].median())
df['Assignments_Avg'] = df['Assignments_Avg'].fillna(df['Assignments_Avg'].median())
```

```python
df.isna().sum()
```

```python
#filling missing values in categorical data with mode 
df['Parent_Education_Level'] = df['Parent_Education_Level'].fillna(df['Parent_Education_Level'].mode()[0])
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 12))
for i, column in enumerate(df.select_dtypes(include=['float64', 'int64']).columns, 1):
    plt.subplot(4, 3, i)
    sns.histplot(df[column], kde=True)
    plt.title(f"Distribution of {column}")
plt.tight_layout()
plt.show()
```

### Distribution Analysis  
- This visualization shows how different numerical features are distributed.
- It helps us detect outliers and understand the overall data spread. 

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df._get_numeric_data().corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Correlation Heatmap  
- This heatmap displays the correlation between different variables.  
- A higher correlation suggests a stronger relationship between two factors.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[["Midterm_Score", "Final_Score", "Total_Score"]])
plt.title("Boxplot of Scores")
plt.show()
```

### Boxplot Analysis  
- Boxplots highlight potential outliers in students' scores.  
- Outliers can indicate exceptional or struggling students.  

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Attendance (%)"].sample(2000), y=df["Total_Score"].sample(2000))
plt.title("Attendance vs Total Score")
plt.show()
```

### Attendance vs. Performance  
- This scatterplot shows if higher attendance leads to better scores.  
- We expect a positive correlation between attendance and performance.  

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Study_Hours_per_Week"].sample(3000), y=df["Total_Score"].sample(3000))
plt.title("Study Hours vs Total Score")
plt.show()
```
