### Data Cleaning 

```python
# Check missing Values 
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Drop rows with missing CustomerID 
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df_original = df.copy()
```

```python
# df (returns) for returns products 
returns = df[df["Quantity"] < 0]  
returns
```

```python
free_items = df[df['UnitPrice'] == 0]
free_items 
```

```python
# Remove negative quantities and prices --> Remove outliers 
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df
```

```python
# convert date to datetime format 
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"] , errors='coerce')
df
```

```python
df = df.dropna(subset=["InvoiceDate"])
df
```

### Feature Engineering 

```python
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)
```

```python
# add column of TotalPrice
df.loc[:, "TotalPrice"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
# add column month
df['Month'] = df['InvoiceDate'].dt.month
df
```

```python
# add column Hour and DayOfWeek to DataFrame
df["DayOfWeek"] = df["InvoiceDate"].dt.day_name()
df["Hour"] = df["InvoiceDate"].dt.hour
df
```

```python
#filter sales that have discounts --> 
discount_sales = df[df["UnitPrice"] < df["UnitPrice"].mean()]
#calculate total revenue per month 
discount_saless = discount_sales.groupby("Month")["TotalPrice"].sum()
total_sales = df.groupby("Month")["TotalPrice"].sum()
```

### Impact of Discounts on Sales

#### Overview  
The line chart displays the impact of discounted sales on total sales over time.  

- **Blue Line --> Total Sales**: Represents the total revenue generated each month.  
- **Orange Dashed Line --> Discounted Sales**: Represents the revenue from products sold at a discount.  

#### Key Insights  
1. **Steady Growth**: Both total and discounted sales show an increasing trend over time.  
2. **Strong Correlation**: As discounted sales increase, total sales tend to rise, suggesting that discounts drive more revenue.  
3. **Peak in Q4 (fourth quarter of the year)**: The highest sales occur in **September, October, and November**, indicating a strong seasonal effect, possibly due to holidays or promotions.  
4. **End-of-Year Drop**: A sharp decline in December, which could indicate a post-holiday slowdown.

```python
import streamlit as st
import plotly.express as px
from streamlit_folium import st_folium


top_countries = df.groupby('Country', as_index=False)['TotalPrice'].sum()

def create_map():
    m = folium.Map(location=[20, 0], zoom_start=2, tiles='CartoDB dark_matter')
    for _, row in top_countries.iterrows():
        folium.CircleMarker(
            location=[51.5, -0.1],
            radius=10,
            popup=f"{row['Country']}: {row['TotalPrice']:,}",
            color='blue',
            fill=True,
            fill_color='blue'
        ).add_to(m)
    return m

st.set_page_config(page_title="Sales Dashboard", layout="wide")
st.sidebar.title("Filter Options")

selected_countries = st.sidebar.multiselect("Select Countries", list(top_countries['Country'].unique()), default=["United Kingdom"])
df_filtered = df[df['Country'].isin(selected_countries)]

start_date, end_date = st.sidebar.date_input("Select Date Range", [df['InvoiceDate'].min().date(), df['InvoiceDate'].max().date()])
df_filtered = df_filtered[
    (df_filtered['InvoiceDate'] >= pd.Timestamp(start_date)) &
    (df_filtered['InvoiceDate'] <= pd.Timestamp(end_date))
]

st.title("Sales Dashboard")
col1, col2, col3 = st.columns(3)
col1.metric("Total Sales", f"{df_filtered['TotalPrice'].sum():,.2f}")
col2.metric("Average Sales", f"{df_filtered['TotalPrice'].mean():,.2f}")
col3.metric("Max Sales", f"{df_filtered['TotalPrice'].max():,.2f}")

# Visualizations
col1, col2 = st.columns(2)

# Bar Chart
top_countries_filtered = df_filtered.groupby('Country', as_index=False)['TotalPrice'].sum()
fig1 = px.bar(top_countries_filtered, x='Country', y='TotalPrice', title="Top Countries by Sales", template='plotly_dark')
col1.plotly_chart(fig1, use_container_width=True)

# Pie Chart
fig2 = px.pie(top_countries_filtered, names='Country', values='TotalPrice', title="Sales Distribution by Country", template='plotly_dark')
col2.plotly_chart(fig2, use_container_width=True)

# Line Chart
monthly_sales_filtered = df_filtered.groupby('Month', as_index=False)['TotalPrice'].sum()
monthly_sales_filtered['Month'] = monthly_sales_filtered['Month'].astype(str)  # تحويل Month إلى نص
fig3 = px.line(monthly_sales_filtered, x='Month', y='TotalPrice', title="Sales Trend Over Time", template='plotly_dark')
st.plotly_chart(fig3, use_container_width=True)

# Scatter Plot
fig4 = px.scatter(df_filtered, x='UnitPrice', y='TotalPrice', title="Sales vs Unit Price", template='plotly_dark')
st.plotly_chart(fig4, use_container_width=True)

# Box Plot
fig5 = px.box(df_filtered, x='Country', y='TotalPrice', title="Sales Distribution by Country", template='plotly_dark')
st.plotly_chart(fig5, use_container_width=True)

# Hourly Sales Analysis
hourly_sales = df_filtered.groupby('Hour', as_index=False)['TotalPrice'].sum()
fig6 = px.bar(hourly_sales, x='Hour', y='TotalPrice', title="Sales by Hour", template='plotly_dark')
st.plotly_chart(fig6, use_container_width=True)

# Map
st.subheader("Sales Map")
st_folium(create_map())
```
