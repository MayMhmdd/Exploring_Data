```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_csv('C:/Users/Royaa/Downloads/Matplotlib Assignment/Titanic.csv')
df
```

```python
sex_counts = df["Sex"].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct="%1.1f%%", colors=["lightblue", "pink"])
plt.title("Distribution of passengers by gender")
plt.show()
```

```python
plt.hist(df["Age"].dropna(), bins=20, color="blue", edgecolor="black")
plt.title("Distribution of passengers by age")
plt.xlabel("age")
plt.ylabel("number of passengers")
plt.show()
```

```python
plt.barh(["die", "survive"],df["Survived"].value_counts() , color=["olive", "gray"] , height = 0.5)
plt.title("number of survivors and non-survivors")
plt.ylabel("number of passengers")
plt.show()
```

```python
plt.scatter(df["Age"], df["Fare"], color="purple")
plt.title("relation between age and ticket price")
plt.xlabel("age")
plt.ylabel("ticket price")
plt.show()
```

```python
fare_by_class = df.groupby("Pclass")["Fare"].mean()
plt.plot(fare_by_class.index, fare_by_class.values, marker="s", color="green")
plt.title("average fare for each class of ticket")
plt.xlabel("ticket class")
plt.ylabel("average ticket price")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

plt.subplot(1,2,1)
plt.hist(df[df['Survived'] == 1]['Age'].dropna(), bins=20, color='skyblue')
plt.title("Age Distribution for Survived")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.subplot(1,2,2)
plt.hist(df[df['Survived'] == 0]['Age'].dropna(), bins=20, color='blue')
plt.title("Age Distribution for Not Survived")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.suptitle("Age Distribution by Survival Status", fontsize=16)
plt.show()
```

```python
survival_rate = df.groupby('Pclass')['Survived'].mean()

plt.pie(survival_rate, labels=['1st Class', '2nd Class', '3rd Class'], autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'salmon'])
plt.title("Survival Rate by Ticket Class")
plt.show()
```

```python
survived_ages = df[df['Survived'] == 1]['Age'].dropna()
not_survived_ages = df[df['Survived'] == 0]['Age'].dropna()

plt.boxplot([survived_ages, not_survived_ages], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='black'),flierprops=dict(markerfacecolor='red', marker='o'))

plt.title("Relation between age and Survival Status ")
plt.xlabel("Survival Status")
plt.ylabel("Age")
plt.xticks([1, 2], ['Survived', 'Not Survived'])
plt.show()
```
